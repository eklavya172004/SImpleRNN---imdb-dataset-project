```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[
    "this movie is great and entertaining",
    "this movie is phenomenol and mustwatch",
    "the plot was boring and predictable",
    "a masterpiece in cinema",
    "waste of time and money",
    "highly recommend this movie",
    "worst film i have ever seen",
    "wonderful story and great direction"
]
```

```python
vocab_size = 10000
```

```python
one_hot_en = [one_hot(words,vocab_size) for words in sent]

one_hot_en
```

```python
from tensorflow.keras.layers import  Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
set_length=8
```

```python
embedded = pad_sequences(one_hot_en,padding='pre',maxlen=set_length)

embedded
```

```python
# Feature representation
dim=10
```

```python
model=Sequential()

model.add(Embedding(vocab_size,dim,input_length=set_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded)
```
